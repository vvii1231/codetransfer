```python
import pandas as pd
import numpy as np


df_is2 = pd.read_csv(r"F:\nasa\is2\atl12\beams\monthly_avg_ssha_gt2r.csv")
df_s3a = pd.read_csv(r"F:\esa\average_monthly_s3asshadtu21.csv")
df_s3b = pd.read_csv(r"F:\esa\average_monthly_s3bsshadtu21.csv")


df_is2['Date'] = pd.to_datetime(df_is2['Date'])
df_s3a['Date'] = pd.to_datetime(df_s3a['Date'])
df_s3b['Date'] = pd.to_datetime(df_s3b['Date'])
```

```python
df_is2.set_index('Date', inplace=True)
df_s3a.set_index('Date', inplace=True)
df_s3b.set_index('Date', inplace=True)


df_diff_s3a = df_is2.subtract(df_s3a, fill_value=0)
df_diff_s3b = df_is2.subtract(df_s3b, fill_value=0)

mean_diff_s3a = df_diff_s3a.mean(axis=1)
mean_diff_s3b = df_diff_s3b.mean(axis=1)


df_final = pd.DataFrame(index=df_is2.index)
df_final['IS-2_SSHA_MEAN'] = df_is2.iloc[:, 0]
df_final['S3A_SSHA_MEAN'] = df_s3a.iloc[:, 0]
df_final['S3B_SSHA_MEAN'] = df_s3b.iloc[:, 0]
df_final['MEAN_DIFF_S3A'] = mean_diff_s3a
df_final['MEAN_DIFF_S3B'] = mean_diff_s3b

# 计算标准差
df_final['SD_DIFF_S3A'] = df_final.apply(lambda x: np.std([x['IS-2_SSHA_MEAN'], x['S3A_SSHA_MEAN']]), axis=1)
df_final['SD_DIFF_S3B'] = df_final.apply(lambda x: np.std([x['IS-2_SSHA_MEAN'], x['S3B_SSHA_MEAN']]), axis=1)

# 添加包含均值和标准差的列
df_final['MEAN(SD)_OF_DIFFERENCES_IS2_S3A'] = df_final.apply(
    lambda x: f"{x['MEAN_DIFF_S3A']:.3f} ({x['SD_DIFF_S3A']:.3f})", axis=1)
df_final['MEAN(SD)_OF_DIFFERENCES_IS2_S3B'] = df_final.apply(
    lambda x: f"{x['MEAN_DIFF_S3B']:.3f} ({x['SD_DIFF_S3B']:.3f})", axis=1)

# 选择最终要b显示的列
final_columns = ['IS-2_SSHA_MEAN', 'S3A_SSHA_MEAN', 'S3B_SSHA_MEAN',
                 'MEAN(SD)_OF_DIFFERENCES_IS2_S3A', 'MEAN(SD)_OF_DIFFERENCES_IS2_S3B']
df_final = df_final[final_columns].reset_index()
```
